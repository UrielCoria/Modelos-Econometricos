```python
from statsmodels.compat import lzip
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

<p style="text-align:center">
    <font face="times new roman" size="8" style="color:darkred">
        <div style="text-align:center">
        Modelo Básico de Consumo Keynesiano
        </div>
    </font>

```python
data = pd.read_csv('../data/Consumo Keynesiano.csv', parse_dates=['Periodos'], index_col='Periodos', infer_datetime_format=True)
data.head()
```

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        Los datos antes mostrados fueron tomados del Banco de Información Económica 
        (<a href="https://www.inegi.org.mx/app/indicadores/?tm=0#bodydataExplorer">BIE</a>) perteneciente al INEGI. Se tomaron las series desestacionalizadas a precios constantes del Consumo Privado
        y el Producto Interno Bruto. Del Sistema de Administración Tributaria se tomó la recaudación anual correspondiente al Impuesto Sobre la Renta, el Impuesto el Valor Añadido y el Impuesto 
        Especial sobre Producción y Servicios y se sumaron para formar un dato trimestral.
        </div>
    </font>

```python
plt.figure(figsize=[16,7])

plt.plot(data['C'], label='Consumo Privado')
plt.plot(data['Y'], label='Producto Interno Bruto')

plt.ylabel('Millones de pesos', size=16)

plt.grid(True)
plt.legend()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()]) # Elimina la escala de notación científica
plt.show()
```

```python
plt.figure(figsize=[16,7])

plt.scatter(data['Y'], data['C'])

plt.xlabel('Producto Interno Bruto', size=16)
plt.ylabel('Consumo', size=16)
plt.title('Correlación Consumo-Ingreso', size=22)

plt.grid(True)
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()])
plt.gca().set_xticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_xticks()])
plt.show()
```

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        La gráfica anterior muestra claramente la relación positiva que existe entre el consumo y el PIB. Por lo tanto, se esperaría que entre más aumente el PIB, existirá mayor consumo en la 
        economía.
        </div>
    </font>

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        Para mostrar dicha relación, se tomará el modelo de consumo keynesiano básico expresado en la siguiente ecuación:<br><br>
        $$ C_{t} = c_{0} + c_{1}(Y_{t}-T) $$
        $$ C_{t} = c_{0} + c_{1}Yd_{t} $$
        </div>
    </font>

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        Dónde:
            <ul>
                <li> $C_{t}$ es el Consumo Privado </li>
                <li> $c_{0}$ es el Consumo autónomo </li>
                <li> $c_{1}$ es la propensión marginal a consumir </li>
                <li> $Yd_{t}$ es el Ingreso disponible medido como el ingreso menos impuestos </li>
                <li> $t$ es el tiempo representado en trimestres </li>
            </ul>
        Los parámetros $c_{0}$ y $c_{1}$ deben ser positivos. Como solo se desea abordar la parte técnca del lenguaje Python, no se explicará la teoría económica detrás de dicha ecuación, por lo
        tanto, se recomienda revisar la sección Consumo (pp. 52-53) del libro Macroeconomia - Teoria y Politica Con Aplicacion En America Latina de los autores Olivier Blanchard y Daniel Perez 
        Enrri.
        </div>
    </font>

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        Se utilizó la serie del Consumo antes mencionada para representar $C_{t}$ en el modelo, para el Ingreso ($Y_{t}$) se tomó el PIB como aproximación y de igual manera los impuestos
        </div>
    </font>

```python
data_niveles = pd.DataFrame(index=data.index)            # Crea un Dataframe para manipular mejor la información sin riesgo a que los datos originales sean modificados
data_niveles['C'] = np.log(data['C'])                    # Se aplica logaritmo a la serie para hacer más homogenea la escala de información y facilidad de interpretación a puntos porcentuales
data_niveles['Yd'] = data['Y'] - data['T']               # Cálculo del Ingreso disponible
data_niveles['Yd'] = np.log(data_niveles['Yd'])          # Aplica logaritmo al Ingreso disponible
Exog = sm.add_constant(data_niveles[['Yd']])  # Se extraen las variables exógenas y se les añade una constante
```

```python
model = sm.OLS(data_niveles['C'], Exog.iloc[:,:2]).fit() # Primer Modelo
model.summary()                                          # Visualiza el modelo
```

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        El modelo anterior es fiel a la ecuación que se desea estimar, es decir, solo contempla el Ingreso disponible. La constante y el ingreso resulta estadísticamente significativo al tener una 
        probabilidad menor a 0.05 y determina que el valor de los coeficientes es diferente de cero.
        </div>
    </font>

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        Con respecto a la normalidad de los residuos se puede ver que tiene una Kurtosis de 2.6, lo cual es muy cercano a la normalidad con valor de 3 y es reforzado con los trabajos de D’Agostino
        y Pearson en los trabajos "Tests for departure from normality" y "An omnibus test of normality for moderate and large sample size". <br>
        La hipótesis nula dice que los residuos vienen de una distribución normal, por lo tanto, una probabilidad menor a 0.05 permite concluir que los residuos no provienen de una distribución
        normal. En este caso es lo contrario, la prueba rechaza la hipótesis nula y se acepta la normalidad de los residuos. La prueba Jarque-Bera también permite concluir que existe normalidad
        en los residuos.
        </div>
    </font>

```python
lzip(['Estadístico de multiplicador de Lagrange', 'Valor-P de Lagrange' , 'Valor F' , 'Valor-P de F'], \
     sm.stats.diagnostic.acorr_breusch_godfrey(model)) # Prueba de Autocorrelación Breusch-Godfrey
```

```python
lzip(['Estadístico del multiplicador de Lagrange', 'Valor-P de Lagrange', 'Valor F', 'Valor-P de F'], \
     sm.stats.diagnostic.het_breuschpagan(model.resid, model.model.exog)) # Prueba de Heterocedasticidad Breusch-Pagan-Godfrey
```

```python
lzip(['Valor F', 'Valor-P de F'], sm.stats.diagnostic.het_goldfeldquandt(model.resid, model.model.exog)) # Prueba de Heterocedasticidad Goldfeld-Quandt
```

```python
lzip(['Estadístico del multiplicador de Lagrange', 'Valor-P de Lagrange', 'Valor F', 'Valor-P de F'], \
     sm.stats.diagnostic.het_white(model.resid, model.model.exog)) # Prueba de Heterocedasticidad White
```

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        Para verificar la existencia de Homocedasticidad se usan las pruebas Breusch-Godfrey, Breusch-Pagan-Godfrey y White, en dos de las tres pruebas se comprueba la Homocedasticidad del modelo y
        resulta acorde a la distribución normal de los residuos.<br>
        En el caso de la autocorrelación se tiene un valor de Durbin-Watson de 0.537, por su distancia al valor 2, indica que existe autocorrelación en el modelo, para asegurar dicha afirmación se
        realizó la prueba Breusch-Godfrey para autocorrelación y se muestra una probabilidad menor a 0.05, por lo tanto, se concluye la existencia de autocorrelación.
        Como el modelo solo tiene una variable, se asume que no puede existir multicolinealidad, pero la salida de Python muestra que existe problemas numéricos en el modelo, uno de ellos podría ser
        multicolinealidad, pero como se mencionó, esto no tendría sentido.
        </div>
    </font>

```python
data_niveles = pd.DataFrame(index=data.index)            # Crea un Dataframe para manipular mejor la información sin riesgo a que los datos originales sean modificados
data_niveles['C'] = np.log(data['C'])                    # Se aplica logaritmo a la serie para hacer más homogenea la escala de información y facilidad de interpretación a puntos porcentuales
data_niveles['Yd'] = data['Y'] - data['T']               # Cálculo del Ingreso disponible
data_niveles['Yd'] = np.log(data_niveles['Yd'])          # Aplica logaritmo al Ingreso disponible
data_niveles['Yd1'] = data_niveles['Yd'].shift()         # Crea un rezago a la serie del Ingreso
data_niveles['C1'] = data_niveles['C'].shift()           # Crea un rezago a la serie del Consumo
data_niveles = data_niveles.dropna()                     # Al obtener rezagos de una serie se pierde la misma cantidad de observaciones en la parte inicial o final de dicha serie
Exog = sm.add_constant(data_niveles[['Yd','Yd1','C1']])  # Se extraen las variables exógenas y se les añade una constante
```

```python
model = sm.OLS(data_niveles['C'], Exog).fit()  # Segundo modelo
model.summary()
```

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        Para corregir los problemas de autocorrelación del primer modelo se agregaron a las variables explicativas rezago en el consumo y otro en el ingreso.<br>
        Al realizar el segundo modelo se obtuvo que las nuevas variables exógenas son estadísticamente significativas individualmente y en su conjunto. Se mantiene la normalidad de los residuos
        de acuerdo a las pruebas Jarque-Bera y Omnibus.
        </div>
    </font>

```python
lzip(['Estadístico de multiplicador de Lagrange', 'Valor-P de Lagrange' , 'Valor F' , 'Valor-P de F'], \
     sm.stats.diagnostic.acorr_breusch_godfrey(model, nlags=1)) # Prueba de Autocorrelación Breusch Godfrey
```

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        La prueba de arriba muestra que no existe autocorrelación en los residuos y el valor Durbin-Watson tiene un valor de 1.8 que es muy cercano al 2. Por lo anterior, se afirma que no existe 
        autocorrelación en los residuos.
        </div>
    </font>

```python
lzip(['Estadístico del multiplicador de Lagrange', 'Valor-P de Lagrange', 'Valor F', 'Valor-P de F'], \
     sm.stats.diagnostic.het_breuschpagan(model.resid, model.model.exog))
```

```python
lzip(['Valor F', 'Valor-P de F'], sm.stats.diagnostic.het_goldfeldquandt(model.resid, model.model.exog))
```

```python
lzip(['Estadístico del multiplicador de Lagrange', 'Valor-P de Lagrange', 'Valor F', 'Valor-P de F'], sm.stats.diagnostic.het_white(model.resid, model.model.exog))
```

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        Las tres pruebas de arriba corresponden a verificar si existe Homocedasticidad en el modelo, de igual manera, dos de las tres pruebas determinan la existencia de Homocedasticidad en el
        modelo.
        </div>
    </font>

```python
VIF = pd.DataFrame()
VIF['Variable'] = Exog.columns[1:]
  
VIF['VIF'] = [variance_inflation_factor(Exog.values[1:], i+1) for i in range(len(Exog.columns[1:]))]
  
VIF
```

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        Lo que se muestra arriba es el VIF para cada variable exógena. Como en el nuevo modelo se están tomando tres variables, sería adecuado revisar la existencia de multicolinealidad mediante el
        VIF. Se observa un valor elevado entre las variables rezagadas, lo cual, tiene sentido por la relación estrecha que tienen ambas variables en el gráfico de correlación.
        </div>
    </font>

```python
plt.figure(figsize=[16,7])

plt.plot(np.exp(data_niveles.C), 'o', label='Consumo')
plt.plot(np.exp(model.fittedvalues), color='black', linestyle='dashed', marker='<', label='Consumo Estimado')
plt.plot(np.exp(model.get_prediction().summary_frame()['obs_ci_lower']), "r--", label='Intervalos de confianza')
plt.plot(np.exp(model.get_prediction().summary_frame()['obs_ci_upper']), "r--")

plt.legend()
plt.grid(True)
plt.title('Modelo', size=22)
plt.ylabel('Millones de pesos', size=16)
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()])

plt.show()
```

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        La gráfica muestra los datos del consumo y el consumo estimado con el modelo y se observa un ajuste cercano entre el modelo y los valores reales. El principal inconveniente es el cambio que
        provocó la pandemia en los periodos posteriores por la fuerte incertidumbre que ha existido en los últimos dos años, por lo cual, es normal que el modelo ya no se ajuste a la realidad y 
        por esa razón el consumo en el periodo de la pandemia sale de los intervalos de confianza.
        </div>
    </font>

```python
fig, axes = plt.subplots(2, figsize=(16, 7))

ax = axes[0]
ax.hist(model.resid)
ax.set(title='Residuos del modelo')
ax.grid(True)


ax = axes[1]
ax.plot(model.resid, label='Residuos del modelo', linestyle='-.')
ax.hlines(y=0, xmin=Exog.index[0], xmax=Exog.index[-1], color='black')
ax.grid(True)
ax.legend()

fig.tight_layout()
```

<p style="text-align:justify">
    <font face="times new roman" size="4">
        <div style="text-align:left">
        Al estimar el modelo se modificó la estructura de la ecuación antes presentada, se le añadieron dos rezagos que antes no se encontraban representados, entonces, la nueva ecuación del modelo
        es la siguiente:<br><br>
        $$C_{t} = c_{0} + c_{1}Yd_{t}+c_{2}Yd_{t-1}+c_{3}C_{t-1}$$<br>
        Y los multiplicadores están dados por los coeficientes del modelo de la siguiente manera:<br><br>
        $$C_{t}=-1.5639 + 1.0837 Yd_{t}-0.7246Yd_{t-1}+0.7289C_{t-1}$$<br>
        Al existir rezagos en un modelo se debe tener en cuenta que el modelo ya no es estático como lo propone el libro, es decir, el modelo se vuelve dinámico y se deben considerar los tiempos de
        ajuste. Lo anterior pertenece a las series temporales y no compete a la intención de este código. En ambos casos se puede revisar la teoría del consumo e ingreso permanente para una 
        mejor comprensión económica.
        </div>
    </font>

```python
data = pd.read_csv('../data/Consumo Indices.csv', parse_dates=['Periodos'], index_col='Periodos', infer_datetime_format=True)
data.head()
```

```python
data_niveles = pd.DataFrame(index=data.index)
data_niveles['C'] = data['C']
data_niveles['Y'] = data['Y']
data_niveles['C'] = data_niveles['C'].pct_change(12)
data_niveles['Y'] = data_niveles['Y'].pct_change(12)
#data_niveles['Yd'] = np.log(data_niveles['Yd'])
data_niveles['Y1'] = data_niveles['Y'].shift()
data_niveles['C1'] = data_niveles['C'].shift()
data_niveles = data_niveles.dropna()
Exog = sm.add_constant(data_niveles[['Y','Y1','C1']])
```

```python
data_niveles.head()
```

```python
model = sm.OLS(data_niveles['C'], Exog).fit()  # Segundo modelo
model.summary()
```

```python
lzip(['Estadístico de multiplicador de Lagrange', 'Valor-P de Lagrange' , 'Valor F' , 'Valor-P de F'], \
     sm.stats.diagnostic.acorr_breusch_godfrey(model, nlags=1)) # Prueba de Autocorrelación Breusch Godfrey
```

```python
lzip(['Estadístico del multiplicador de Lagrange', 'Valor-P de Lagrange', 'Valor F', 'Valor-P de F'], \
     sm.stats.diagnostic.het_breuschpagan(model.resid, model.model.exog))
```

```python
lzip(['Valor F', 'Valor-P de F'], sm.stats.diagnostic.het_goldfeldquandt(model.resid, model.model.exog))
```

```python
lzip(['Estadístico del multiplicador de Lagrange', 'Valor-P de Lagrange', 'Valor F', 'Valor-P de F'], \
     sm.stats.diagnostic.het_white(model.resid, model.model.exog))
```

```python
VIF = pd.DataFrame()
VIF['Variable'] = Exog.columns[1:]
  
VIF['VIF'] = [variance_inflation_factor(Exog.values[1:], i+1) for i in range(len(Exog.columns[1:]))]
  
VIF
```

```python
plt.figure(figsize=[16,7])

plt.plot(np.exp(data_niveles.C), 'o', label='Consumo')
plt.plot(np.exp(model.fittedvalues), color='black', linestyle='dashed', marker='<', label='Consumo Estimado')
plt.plot(np.exp(model.get_prediction().summary_frame()['obs_ci_lower']), "r--", label='Intervalos de confianza')
plt.plot(np.exp(model.get_prediction().summary_frame()['obs_ci_upper']), "r--")

plt.legend()
plt.grid(True)
plt.title('Modelo', size=22)
plt.ylabel('Millones de pesos', size=16)
#plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()])

plt.show()
```
